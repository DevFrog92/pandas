```python
student1 = pd.Series({
  '국어': np.nan,
  '수학': 50,
  '영어': 80
})

student1
```

```python
percentage = student1 / 200
percentage
```

```python
student2 = pd.Series({
  '수학': 100,
  '국어': 50,
})

student2
```

```python
addition = student1 + student2
subtraction = student1 - student2
multiplication = student1 * student2
division = student1 / student2

print(type(division))
print(addition)
print(subtraction)
print(multiplication)
print(division)
```

```python
pd.DataFrame([addition, subtraction, multiplication, division], index=['덧셈','뺄셈','곱셈','나눗셈'])
```

```python
add_sr = student1.add(student2, fill_value=0)
sub_sr = student1.sub(student2, fill_value=0)
mul_sr = student1.mul(student2, fill_value=0)
div_sr = student1.div(student2, fill_value=0)
```

```python
pd.DataFrame([add_sr, sub_sr, mul_sr, div_sr], index=['덧셈','뺄셈','곱셈','나눗셈'])
```
